```python
file2 = r'/home/erick/baseballdatabank-master/core/People.csv'
```

```python
players = pd.read_csv(file2)
```

```python
bigGameBats = postBats[postBats["round"] == 'WS']
```

```python
bigGameBats.playerID
```

```python
file3 = r'/home/erick/baseballdatabank-master/core/Teams.csv'
```

```python
players.nameGiven = players['nameFirst']+'_'+players['nameLast']
```

```python
players.head()
```

```python
teams = pd.read_csv(file3)
```

```python
teams.head()
```

```python
players_noDups = players.drop_duplicates()
```

```python
players.groupby("teamID").playerID.count()
```

```python
playerBats = pd.merge(bigGameBats, players_noDups[["nameGiven", "playerID"]], on="playerID")
```

```python
teamPlayerBats = pd.merge(playerBats, teams[["name", "teamID"]], on="teamID")
```

```python
tpb_nodups = teamPlayerBats.drop_duplicates(subset=["yearID", "playerID", "teamID"])
```

```python
result = tpb_nodups.groupby(["teamID", "yearID"]).playerID.count()
```

```python
result.plot.bar(x='teamID', y='yearID', rot=0)
```
